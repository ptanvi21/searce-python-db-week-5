```python
# an HTTP request in python
import socket
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect(('data.pr4e.org',80))
c = 'GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
s.send(c)

while True:
    data = s.recv(512)
    if(len(data) < 1):
        break
    print(data.decode())
s.close()
```

```python
import urllib.request, urllib.parse, urllib.error
import json
serviceurl = 'http://py4e-data.dr-chuck.net/json?'
while True:
    add = "Pune"
    if len(add) < 1: break

    url = serviceurl + urllib.parse.urlencode({'address': add})

    print('Retrieving', url)
    uh = urllib.request.urlopen(url)
    data = uh.read().decode()
    print('Retrieved', len(data), 'characters')

    try:
        js = json.loads(data)
    except:
        js = None

    if not js or 'status' not in js or js['status'] != 'OK':
        print('Failed To Retrieve')
        print(data)
        continue

    lat = js['results'][0]['geometry']['location']['lat']
    lng = js['results'][0]['geometry']['location']['lng']
    print('latitude - ', lat, 'longitude - ', lng)
    location = js['results'][0]['formatted_address']
    print(location)
```

```python
# Inheritance
class Animal:
    x = 0
    name = ""

    def __init__(self,a):
        self.name = a
        print(self.name, "constructed")

    def party(self):
        self.x = self.x + 1
        print(self.name,"count is",self.x)

class Fan(Animal):
    y = 0
    def count(self):
        self.y = self.y + 5
        self.party()
        print(self.name,"points are",self.y)

s = Animal('Monkey')
s.party()
j = Fan("Dog")
j.party()
j.count()
```

```python
import sqlite3

conn = sqlite3.connect('emaildb.sqlite')
cur = conn.cursor()

cur.execute('DROP TABLE IF EXISTS Counts')

cur.execute('''
CREATE TABLE Counts (email TEXT, count INTEGER)''')

fname = input('Enter file name: ')
if (len(fname) < 1): fname = 'sample.txt'
fh = open(fname)
for line in fh:
    if not line.startswith('From: '): continue
    pieces = line.split()
    email = pieces[1]
    cur.execute('SELECT count FROM Counts WHERE email = ? ', (email,))
    row = cur.fetchone()
    if row is None:
        cur.execute('''INSERT INTO Counts (email, count)
                VALUES (?, 1)''', (email,))
    else:
        cur.execute('UPDATE Counts SET count = count + 1 WHERE email = ?',
                    (email,))
    conn.commit()

sqlstr = 'SELECT email, count FROM Counts ORDER BY count DESC LIMIT 10'

for row in cur.execute(sqlstr):
    print(str(row[0]), row[1])

cur.close()
```

```python
# multiplication

def multi(a,b):
    if(b==1):
        return a
    else:
        return a+multi(a,b-1)

prod = multi(10,5)
print(prod)
```

```python
# factorial
def fact(n):
    if(n==1):
        return 1
    else:
        return n*fact(n-1)
print(fact(4))
```

```python
# fibonacci
def fib(n): 
    if(n==0 or n==1):
        return 1
    else:
        return fib(n-1)+fib(n-2)
print(fib(10))
```

```python
# Palindrome
def isPalindrome(s):
    t = s[::-1]
    if(s==t):
        return 1
    else:
        return 0
s = input("ENter string")
ans = isPalindrome(s)
if(ans==1):
    print("Given string is palindrome")
else:
    print("Given string is not palindrome")
```
